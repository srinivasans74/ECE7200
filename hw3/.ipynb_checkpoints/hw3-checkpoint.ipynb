```python
#Srinivasan Subramaniyan
import sys
import os
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
s=100
states=np.arange(s)
action=np.arange(s+1)
rw=0
count=0
epsilon=0.0001

def vi(ph):
    v = [0 for _ in range(101)]
    r = [0 for _ in range(101)]
    r[-1]=1
    pi=[0 for _ in range(101)]
    v[-1]=1
    delta=1
    a=100;
    while delta > epsilon:
        delta = 0
        for i in range(1,s):  
            vinit = v[i]
            posi = min(i, a - i)
            q = np.zeros(posi)
            for ai in range(1, posi+1):
                q[ai - 1] = ((ph* v[i + ai])+r[i]) + (((1 -ph) * v[i - ai])+r[i])
            v[i] = np.max(q)
            pi[i] = np.argmax(q)+ 1 
            delta = max(delta, abs(vinit - v[i]))
    return pi,v

pi,via=vi(0.25)
plt.plot(action,pi)
plt.ylabel('Final Policy Stake')
plt.xlabel('Capital')
plt.title('0.25')
pi,via=vi(0.55)
#prnt()
fig,ax=plt.subplots()
plt.plot(action,pi)
plt.ylabel('Final Policy Stake')
plt.xlabel('Capital')
plt.title('0.55')
```

```python
vbackup = via
V = [0 for _ in range(101)]
r = [0 for _ in range(101)]
r[-1]=1 #only terminal reward
Returns = [[] for _ in range(101)]
    
def generate_episode(pi, p_h):
    S = []
    A = []
    R = []
    currentstate = random.randint(1,100)
    while (currentstate not in [0,100]):
        S.append(currentstate)
        A.append(pi[currentstate])
        toss = random.random()
        if toss < p_h:   #if we win
            next_state = currentstate + pi[currentstate]
        else:
            next_state = currentstate - pi[currentstate]
        currentstate = next_state
        R.append(r[currentstate])
    return A, S, R


delta_list = []
for counter in (range(10000)):
    delta = 0
    V_old = V
    A, S, R = generate_episode(pi, 0.55)
    G = 0
    for t in list(range(len(S)-1,-1,-1)):
        G = G + R[t]
        Returns[S[t]].append(G)
     #   V[S[t]] = np.average(Returns[S[t]])
    #delta_list.append(np.abs(np.sum(V)-np.sum(V_old)))

#i am doing this outside to reduce computration overhead
for i in range(101):
    V[i]=np.mean(Returns[i])

plt.plot(V)

#plt.plot(delta_list)
plt.ylabel('Value estimate')
plt.xlabel('Capital')
```
